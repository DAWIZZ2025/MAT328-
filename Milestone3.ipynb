```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
menu = pd.read_csv("starbucks_drinkMenu_expanded.csv")
menu
menu.columns = ['Beverage_category','Beverage','Beverage_prep','Calories','Total Fat', 'Trans Fat', 'Saturated Fat', 'Sodium', 'Total Carbohydrates', 'Cholesterol', 'Dietary Fibre', 'Sugars', 'Protein', 'Vitamin A', 'Vitamin C', 'Calcium', 'Iron','Caffeine']
```

```python
menu
```

```python
menu["n"] = menu["Vitamin A"].str.replace("%","")
menu["o"] = menu["Vitamin C"].str.replace("%","")
menu["p"] = menu["Calcium"].str.replace("%","")
menu["q"] = menu["Iron"].str.replace("%","")
menu["r"] = menu["Caffeine"].str.replace("%","")
```

```python
d_filter = menu["Calories"]
d_filter.hist()
plt.title('Sodium vs. Calories in Starbucks Drinks')
plt.xlabel('Calories')
plt.ylabel('Sodium')
```

```python
m_filter = menu[["Protein", "Sugars"]]
m_filter.plot(kind = "line")
plt.title('Protein Count vs. Sugar Count of Starbucks Drinks')
```
